```python
import random
st = int(input('학생 수:'))
res = [None] * st

for i in range(st):
    res[i] = random.randint(1, 100)

total = 0
for i in res:
    total += i

cnt = 0
for i in res:
    cnt = cnt + 1
    print(f'학생{cnt}의 점수: {i}점')
print(f'총점: {total}점')
print(f'평균: {total/st:.2f}점')
```
