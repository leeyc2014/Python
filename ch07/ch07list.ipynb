리스트의 삭제 방법:


```python
# remove 사용(임의의 값을 가지는 요소의 제거)
x = [11, 22, 33, 44, 55, 66, 77]
print(f'x = {x}')
x.remove(33)
print(f'x = {x}')
```

```python
# pop 사용(지정 요소의 제거)
x = [11, 22, 33, 44, 55, 66, 77]
print(f'x = {x}')
v = x.pop(2)
print(f'x = {x}')
print(f'v = {v}')
```

```python
# del문으로 삭제
x = [11, 22, 33, 44, 55, 66, 77]
print(f'x = {x}')
del x[2]
print(f'x = {x}')
del x[2:4]
print(f'x = {x}')
del x[:]
print(f'x = {x}')
```

```python
# 리스트 이터러블
x = ['john', 'george', 'paul', 'ringo']
for i in range(len(x)):
    print(f'x[{i}] = {x[i]}')

print('\n')
for i, name in enumerate(x):        # 많이 사용함: 인덱스, 값
    print(f'x[{i}] = {name}')

print('\n')
for i, name in enumerate(x, 1):
    print(f'x[{i}] = {name}')

print('\n')
cnt = 0
for i in x:
    print(f'x[{cnt}] = {i}')
    cnt += 1
```

```python
# 점수를 입력받아 총점과 평균 구하기 1
import random

print('총점과 평균을 구해봅시다.')
print('----------------------------')
st = int(input('학생 수:'))
res = [None] * st

for i in range(st):
    res[i] = random.randint(1, 100)

total = 0
for i in res:
    total += i

cnt = 0
for i in res:
    cnt = cnt + 1
    print(f'학생{cnt}의 점수: {i}점')
print('----------------------------')
print(f'총점: {total}점')
print(f'평균: {(total/st):.2f}점')
```

```python
# 행렬의 합 구하기(행수/열수/값 입력받기)
import random

print('행렬의 합을 구해봅시다.')
height = int(input('행수:'))
width  = int(input('열수:'))

a = [None] * height     # a[i] 행이 만들어짐
for i in range(height):
    a[i] = [None] * width   # a[i][j] 열이 만들어짐
    for j in range(width):
        a[i][j] = random.randint(1, 10)
        print((f'a[{i}][{j}] : {a[i][j]}'))
print('----------------------------')
print(f'a = {a}')
print('----------------------------')

b = [None] * height
for i in range(height):
    b[i] = [None] * width
    for j in range(width):
        b[i][j] = random.randint(1, 10)
        print((f'b[{i}][{j}] : {b[i][j]}'))
print('----------------------------')
print(f'b = {b}')
print('----------------------------')

c = [None] * height
for i in range(height):
    c[i] = [None] * width
    for j in range(width):
        c[i][j] = a[i][j] + b[i][j]

for i in range(height):
    for j in range(width):
        print(f'c[{i}][{j}] = {c[i][j]}')
print('----------------------------')
print(f'c = {c}')
print('----------------------------')
```

얕은 복사, 깊은 복사(1차원 배열, 2차원 배열): 

```python
x = [[1, 2, 3], [4, 5, 6]]
y = x.copy()
x[0][1] = 9
print(f'x = {x}')
print(f'y = {y}')
```

```python
import copy
x = [[1, 2, 3], [4, 5, 6]]
y = copy.deepcopy(x)
x[0][1] = 9
print(f'x = {x}')
print(f'y = {y}')
```

```python
lst1 = [1, 2, 3, 4, 5]
lst2 = lst1.copy()
lst1[2] = 9
print(f'lst1 = {lst1}')
print(f'lst2 = {lst2}')
```

```python
lst1 = [[1, 2], [3, 4], [5, 6]]
lst2 = lst1.copy()
lst2[0][1] = 9
print(f'lst1 = {lst1}')
print(f'lst2 = {lst2}')
```
