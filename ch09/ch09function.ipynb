```python
# 직각 이등변 삼각형과 사각형

def put_star(n):
    """n개의 '*'을 연속해서 출력"""
    for _ in range(n):
        print('*', end='')

print('직각 이등변 삼각형')
n = int(input('짧은 변의 길이：'))

for i in range(1, n + 1):
    put_star(i)
    print()

print('\n사각형')
w = int(input('가로：'))
h = int(input('세로：'))

for i in range(1, h + 1):
    put_star(w)
    print()
```

```python
# 두 값의 최솟값과 최댓값 구하기

def min_max2(a, b):
    """a와 b의 최솟값과 최댓값을 구하여 리턴"""
    return (a, b) if a < b else (b, a)

n1 = int(input('정수 n1：'))
n2 = int(input('정수 n2：'))

minimum, maximum = min_max2(n1, n2)         # pack한 후에 unpack(리턴값이 여러개면 암묵적으로 튜플로 pack한 후에 unpack)
print('최솟값은', minimum, '이고 최댓값은', maximum, '입니다.')
```

```python
# 1부터 n까지의 합 구하기

def sum_1ton(n):
    """1부터 n까지의 합 구하기"""
    s = 0
    while n > 0:
        s += n
        n -= 1
    return s

x = 100
total = sum_1ton(x)
print('1부터 ', x, '까지의 합은 ', total, '입니다.', sep='')
```

```python
# 리스트의 요소 값 변경하기

def change(lst, idx, val):
    """lst[idx]의 값을 val으로 변경"""
    lst[idx] = val

x = [11, 22, 33, 44, 55]
print('x =', x)
y = x[:]
print(id(x), id(y))
x2 = x
print(id(x), id(x2))
index = 2
value = 66

change(x[:], index, value)
print('x =', x)
change(x, index, value)
print('x =', x)

```

```python
# 직각 이등변 삼격형과 사각형 출력하기2(키워드 인수)

# n개의 s를 연속해서 출력
def puts(n, s): # 위치 인수
    for i in range(n):
        print(s, end='')

print('직각 이등변 삼각형')
n = int(input('짧은 변의 길이：'))

for i in range(1, n + 1):
    puts(n = i, s = '*')
    print()

print('사각형')
w = int(input('가로：'))
h = int(input('세로：'))

for i in range(1, h + 1):
    puts(s = '+', n = w)    # 키워드 인수(함수 호출 시 매개변수명 = 값) - 순서 상관 X(단, 위치 인수와 같이 쓰려면 키워드 인수를 뒤쪽에 둔다)
    print()
```

```python
# 두 개 이상의 값 중 최댓값 구하기

def max2more(a, b, *num):       # *를 붙여 선언된 매개 변수는 암묵적으로 pack을 해서 튜플로 받음
    """두 개 이상의 값 중 최댓값 구하기"""
    max = a if a > b else b
    for n in num:
        if n > max:
            max = n
    return max

print('max2more(1, 2)          = ', max2more(1, 2))
print('max2more(1, 2, 3)       = ', max2more(1, 2, 3))
print('max2more(1, 2, 3, 4, 5) = ', max2more(1, 2, 3, 4, 5))
#print('max2more(1)             = ', max2more(1))        max2more() missing 1 required positional argument: 'b'
```
