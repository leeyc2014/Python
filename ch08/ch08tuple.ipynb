```python
lst = [1, 2, 3]
lst2 = list(range(5))
print(f'lst2 = {lst2}')
lst2[0] = 9     # 가능

tup = (1, 2, 3)
tup2 = tuple(range(5))
print(f'tup2 = {tup2}')
#tup2[0] = 9     # 불가능('tuple' object does not support item assignment)
```

```python
tup1 = tuple([1, 2, 3])
tup2 = tuple({1, 2, 3})
tup3 = tuple((1, 2, 3))
print(f'{tup1}, {tup2}, {tup3}')
```

```python
tup4 = divmod(13, 3)    # 첫 번째 인수를 두 번째 인수로 나눈 몫과 나머지를 튜플로 생성
print(f'tup4 = {tup4}')

div, mod = divmod(13, 3)    # unpack
print(f'div = {div}, mod = {mod}')
```

```python
lst = [1, 2, 3]
print(f'lst = {lst}')
print(f'id = {id(lst)}')
lst += [4, 5]
print(f'lst = {lst}')
print(f'id = {id(lst)}')        # 리스트는 누적 대입 후 제자리 연산으로 아이덴티티가 바뀌지 않음

print('------------------------------')

tpl = (1, 2, 3)
print(f'tpl = {tpl}')
print(f'id = {id(tpl)}')
tpl += (4, 5)
print(f'tpl = {tpl}')
print(f'id = {id(tpl)}')        # 튜플은 아이덴티티가 바뀌며 누적 대입으로 새로운 튜플이 생성됨
```

```python
x = (1, 3, 2)
print(f'x = {x}')
print(f'id = {id(x)}')
x = tuple(sorted(x))            # sorted 함수로 정렬한 것을 리스트로 생성
print(f'x = {x}')               # 생성된 리스트를 튜플로 변환
print(f'id = {id(x)}')
```

```python
tpl1 = (1, 2, 3, 4, 5)
tpl2 = tuple(tpl1)
print(f'tpl1 = {tpl1}')
print(f'tpl2 = {tpl2}')
print(f'tpl1id = {id(tpl1)}')
print(f'tpl2id = {id(tpl2)}')
```

```python
tpl4 = 1, 2, 3, 4       # pack
print(f'tpl4 = {tpl4}')
a, b, c, d = tpl4       # unpack
print(f'a:{a}, b:{b}, c:{c}, d:{d}')
e, _, f, _ = tpl4       # pack을 먼저 하고 unpack
print(f'e:{e}, f:{f}')
e, f = f, e             # 값의 교환에도 튜플 사용(f, e를 평가 -> 그 값을 갖는 튜플 생성 -> 만들어진 튜플 언팩 -> 차례로 e, f에 대입)
print(f'e:{e}, f:{f}')
x, _, *y = tpl4         # *를 사용해 unpack을 리스트로 추출(최대 1개까지 적용)
print(f'x:{x}, y:{y}')
```

```python
s = 'abcde'
for i, ch in enumerate(s):      # 이터러블 객체에서 인덱스와 요소의 쌍으로 pack한 튜플을 추출
    print(f's[{i}] = {ch}')     # 추출된 튜플은 unpack되어 두 변수에 대입
```

```python
x = ['john', 'george', 'paul', 'ringo']
for i, name in enumerate(x):
    print(f'x[{i}] = {name}')
```

```python
# 튜플 리스트(모든 요소가 튜플인 리스트)

students = [
    (2012, '김서준'),
    (2013, '이서아'),
    (2011, '최하준'),
]

print('students =', students)
print('students[0]    =', students[0])
print('students[1]    =', students[1])
print('students[2]    =', students[2])
print('students[0][0] =', students[0][0])
print('students[0][1] =', students[0][1])
print('students[1][0] =', students[1][0])
print('students[1][1] =', students[1][1])
print('students[2][0] =', students[2][0])
print('students[2][1] =', students[2][1])
```

튜플도 리스트와 동일하게 인덱스 연산자 []를 두 번 적용한 인덱스식을 사용


```python
p1 = [1, 2, 3, 4]
p2 = [100, 200, 300, 400]
name = ['a', 'b', 'c', 'd']
z = zip(p1, p2)
print(f'z = {z}')
print(f'z = {list(z)}')
pt = tuple(zip(p1, p2, name))
print(f'pt = {pt}')
```

```python
num = range(6)
print(f'num = {num}')
print(f'num = {list(num)}')
```
