```python
# 계절을 표시하는 사전
# 키는 한국어/값은 영어

season = {
    '봄': 'spring',
    '여름': 'summer',
    '가을': 'autumn',
    '겨울': 'winter',
}

print("season['봄'] = ", season['봄'])
print("season['여름'] = ", season['여름'])
print("season['가을'] = ", season['가을'])
print("season['겨울'] = ", season['겨울'])
```

```python
# 계절을 표시하는 리스트
# 요소는 영어

season = [
    'spring',
    'summer',
    'autumn',
    'winter',
]

print('season[0] = ', season[0])
print('season[1] = ', season[1])
print('season[2] = ', season[2])
print('season[3] = ', season[3])
```

dictionary는 인덱스 값이 key값{'key':'value', ...}, list는 인덱스 값이 정수['value', ...] (list가 key와 value를 가지려면 튜플로 구성된 list([('key', 'value'), ...]))
dict() 함수로 생성                                  list() 함수로 생성

```python
lst = [('a', 1), ('b', 2), ('c', 3), ('d', 4)]
dict4 = dict(lst)
print(f'dict4 = {dict4}')

key = ['a', 'b', 'c', 'd']
value = [1, 2, 3, 4]
dict5 = dict(zip(key, value))
print(f'dict5 = {dict5}')
```

```python
list1 = ['k', 'c', 'j']
list2 = ['k', 'j', 'c']
list1 == list2
```

```python
dict1 = {'k':'kor', 'c':'chn', 'j':'jpn'}
dict2 = {'k':'kor', 'j':'jpn', 'c':'chn'}
dict1 == dict2

#print(dict1[0])       KeyError: 0(인덱싱 사용 불가)
print(dict1['k'])
#print(dict1['u'])     KeyError: 'u'(없는 키 처리 x)
print(dict1.get('u'))       # 없는 키 처리(None)

dict1['u'] = 'usa'          # 없을 경우 값 추가
print(f'dict1 = {dict1}')
dict1['k'] = 'korea'        # 있을 경우 값 수정
print(f'dict1 = {dict1}')
```

dictionary는 요소에 순서가 없기 때문에 순서가 달라도 동일하다고 판정, 슬라이스식은 적용할 수 없음(인덱싱도 사용 불가)

```python
color = {'red': '빨강', 'green': '초록', 'blue': '파랑'}
v1 = color.setdefault('red', '레드')        # 값이 달라도 변경하지 않음
print(f'color = {color}')
print(f'v1 = {v1}')
v2 = color.setdefault('gray', '회색')       # 값이 없으면 추가
print(f'color = {color}')
print(f'v2 = {v2}')

cmy = {'cyan': '하늘', 'magenta': '자주', 'yellow': '노랑'}
color.update(cmy)                           # 기존 dictionary에 결합(중복 값이 있을 경우 제외하고 결합)
print(f'color = {color}')
#color.update('black':'검정')               SyntaxError: invalid syntax
color.update(black='검정')                  
color.update({'purple':'보라'})             # 함수의 매개변수 전달 방식: 키워드 전달 방식
print(f'color = {color}')
```

```python
color = {'red': '빨강', 'green': '초록', 'blue': '파랑', 'cyan': '하늘', 'magenta': '자주', 'yellow': '노랑'}
color2 = color.pop('yellow')
print(f'color = {color}')
print(f'color2 = {color2}')
```

```python
# 딕셔너리의 모든 키를 enumerate 함수로 탐색하기

rgb = {'red': '빨강', 'green': '초록', 'blue': '파랑'}

for i, key in enumerate(rgb):
    print('{} {}'.format(i, key))
print('-------')
for i, key in enumerate(rgb, 1):
    print('{} {}'.format(i, key))
print('-------')
for key in rgb:
    print('{}'.format(key))
```

```python
# 딕셔너리의 모든 키와 값, 요소를 리스트로 추출하기

rgb = {'red': '빨강', 'green': '초록', 'blue': '파랑'}

print('키：', list(rgb.keys()))       # 키의 뷰를 리스트로 변환
print('값：', list(rgb.values()))    # 값의 뷰를 리스트로 변환
print('요소：', list(rgb.items()))     # (키, 값)의 뷰를 리스트로 변환
```

```python
# 딕셔너리에서 추출한 뷰 탐색

rgb = {'red': '빨강', 'green': '초록', 'blue': '파랑'}

# keys()로 추출한 뷰 탐색
for key in rgb.keys():
    print(key)
print('-------')
# values()로 추출한 뷰 탐색
for value in rgb.values():
    print(value)
print('----------------------')
# itmes()로 추출한 뷰 탐색
for item in rgb.items():
    print(item)
```
